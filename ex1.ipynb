```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_excel('customer_churn_large_dataset.xlsx')
```

```python
df=pd.get_dummies(df,columns=['Gender','Location'])
```

```python
df
```

```python
df=df.drop(['CustomerID','Name','Gender_Female'],axis=1)
```

```python

df
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns='Churn')
y = df['Churn']

# splitting for train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.33,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

sc.fit(X_train)

X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
# ------- Part-2: Build the ANN --------
# import keras library and packages
import keras
from keras.models import Sequential  # sequential: acts as a container,that holds all the layers of the ANN together.
from keras.layers import Dense  # an ANN  hidden layer.

#activation funtion : is the function f(x): sigmoid(x)/relu(x)/tanh(x) which is present in each neuron, and is responsible for learning non linear patterns.
#its the function responsible for converting input(x1)-->signal  and passing to the next layer.

#kernel initializer : various ways of initializing the weight matrix.

#optimizer : a function g(x): responsible for reducing the loss, and increasing the accuracy of the model.

#loss : a metric to calculate the error from the NN.

# Initializing the ANN

classifier = Sequential()
# Adding the input layer and the first hidden layer
ip_layer = Dense(activation='relu', input_dim=10, kernel_initializer='uniform', units=100)
classifier.add(ip_layer)  # add  layer to container/sequential


# Adding first hidden layer
layer_info = Dense(activation='relu', kernel_initializer='uniform', units=1000)
classifier.add(layer_info)

layer_info = Dense(activation='relu', kernel_initializer='uniform', units=500)
classifier.add(layer_info)


layer_info = Dense(activation='relu', kernel_initializer='uniform', units=250)
classifier.add(layer_info)

layer_info = Dense(activation='relu', kernel_initializer='uniform', units=100)
classifier.add(layer_info)

# Adding output layer
layer_info = Dense(activation='sigmoid', kernel_initializer='uniform', units=1)
classifier.add(layer_info)

# Compiling the ANN
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

```

```python
classifier.summary()
```

```python
classifier.fit(X_train, y_train,batch_size=256, epochs=25)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier,BaggingClassifier,AdaBoostClassifier,GradientBoostingClassifier,StackingClassifier,VotingClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB,BernoulliNB
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
## Models

lr=LogisticRegression()

dtree=DecisionTreeClassifier()

bagg=BaggingClassifier()

rf=RandomForestClassifier()

knn=KNeighborsClassifier()


ada=AdaBoostClassifier()

gdb=GradientBoostingClassifier()

xgb=XGBClassifier()

naive_gbm=GaussianNB()
naive_bnm=BernoulliNB()
```

```python
models = [('Logistic',lr),('Decision Tree',dtree),('Random Forest',rf),('KNN',knn),('Bagging',bagg),('Adaboost',ada),
          ('Gradient Boosting',gdb),('XGBoost',xgb),('GaussianNB',naive_gbm),('BerboulliNB',naive_bnm)]

evc= VotingClassifier(estimators=models,voting='hard')
```

```python
models = [('Logistic',lr),('Decision Tree',dtree),('Random Forest',rf),('KNN',knn),('Bagging',bagg),('Adaboost',ada),
          ('Gradient Boosting',gdb),('XGBoost',xgb),('GaussianNB',naive_gbm),('BerboulliNB',naive_bnm),('Voting',evc)]
```

```python
for model_name,model in models:
    model.fit(X_train,y_train)
    
    y_train_pred=model.predict(X_train)
    
    train_accuracy=accuracy_score(y_train,y_train_pred)
    
    y_test_pred=model.predict(X_test)
    
    test_accuracy=accuracy_score(y_test,y_test_pred)
    
    print(f"Model: {model_name}")
    print(f"Train_Accuracy: {train_accuracy:.2f}")
    print(f"Test_Accuracy: {test_accuracy:.2f}")
    print('---Confusion Matrix---')
    print(confusion_matrix(y_test , y_test_pred))
    print('---classification report---')
    print(classification_report(y_test , y_test_pred)) 
    print("****************************************************************")
    
    
```

```python
## Cross-validation
from sklearn.model_selection import cross_val_score

for model_name,model in models:
    cv_train_scores = cross_val_score(model,X_train,y_train,cv=20,scoring='accuracy')
    cv_test_scores = cross_val_score(model,X_test,y_test,cv=20,scoring='accuracy')
    
    mean_train_accuracy = cv_train_scores.mean()
    mean_test_accuracy = cv_test_scores.mean()
    
    print(f"Model: {model_name}")
    print(f"Mean_Train_Accuracy: {mean_train_accuracy:.2f}")
    print(f"Mean_Test_Accuracy: {mean_test_accuracy:.2f}")
    print("****************************************************************")
```
